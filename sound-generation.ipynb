```python
!pip install torchsde huggingface_hub
import torch
import soundfile as sf
from diffusers import StableAudioPipeline
from IPython.display import Audio
from google.colab import userdata

# 1. Retrieve the token securely from Colab Secrets
# Make sure you added a secret named 'HF_TOKEN' in the left sidebar (key icon)
hf_token = userdata.get('HF_TOKEN')

# 2. Load the model using the secure token
pipe = StableAudioPipeline.from_pretrained(
    "stabilityai/stable-audio-open-1.0",
    torch_dtype=torch.float16,
    token=hf_token
)
pipe = pipe.to("cuda")

# define the prompts
prompt = "Dolphins speaking to each other, happy, lots of clicks and excitement"
negative_prompt = "Low quality"

# set the seed for generator
generator = torch.Generator("cuda")

# run the generation
audio = pipe(
    prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=200,
    audio_end_in_s=15.0,
    num_waveforms_per_prompt=3,
    generator=generator,
).audios

output = audio[0].T.float().cpu().numpy()
sf.write("output.wav", output, pipe.vae.sampling_rate)

# Play the audio in Colab
Audio("output.wav")
```

```python
import torch
import torchaudio
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write
from IPython.display import Audio

# 1. Load the pre-trained model (small version is faster for Colab)
model = MusicGen.get_pretrained("facebook/musicgen-small")
model.set_generation_params(duration=10)

# 2. Define the prompt
description = "Underwater whale songs with deep resonance and bubbles"

# 3. Generate the audio
print("Generating audio... (this may take a minute)")
wav = model.generate([description])

# 4. Save and Play
# The output filename will be 'whale_sounds.wav'
audio_write("whale_sounds", wav[0].cpu(), model.sample_rate, strategy="loudness")
Audio("whale_sounds.wav")
```
